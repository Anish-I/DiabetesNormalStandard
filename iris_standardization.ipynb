```python
import numpy as np
from sklearn.preprocessing import StandardScaler
# importing the modules
from IPython.display import display
import pandas as pd

# Load a csv file into a dataframe 
df = pd.read_csv ('C:/Users/gaok/OneDrive - Eastern Connecticut State University/Kehan Gao D/ECSU/Fall 2023/CSC305/Examples/iris.csv')

print("Original Data:")
display(df)

# Extract the last column from a dataframe
lastCol=df.iloc[:,-1:]

# Drop the last column from a dataframe
df.drop(columns=df.columns[-1],  axis=1,  inplace=True)

# Convert a dataframe to an array
df2=df.to_numpy()

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data using the scaler
normalized_data = scaler.fit_transform(df2)

# Convert an array back to a dataframe
df = pd.DataFrame(normalized_data, columns =['sepal_length','sepal_width','petal_length','petal_width'])

# Combine the normalized data with the class attribute in a dataframe  
df = pd.concat([df, lastCol], axis=1)

print("\nNormalized Data:")
display(df)

#save a dataframe into a csv file
df.to_csv('iris_standardized.csv')

```
